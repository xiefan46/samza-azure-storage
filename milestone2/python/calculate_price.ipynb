```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
types = ["p4", "p6", "p10", "p15", "p20", "p30", "P40", "P50", "P60", "P70", "P80"]
price = [5.28, 10.21, 19.71, 38.02, 73.22, 135.17, 259.05, 495.57, 946.08, 1802.06, 3604.11]
throughput = [25, 50, 100, 125, 150, 200, 250, 250, 500, 750, 900]

```

```python
dollar_per_100MB = []
for i in range(len(types)):
    dollar_per_100MB.append(int(price[i] * 100 / throughput[i]))
```

```python
print(dollar_per_100MB)
```

```python
def plot_bar(disk_names, results, y_label, title, color):
    y_pos = np.arange(len(disk_names))
    plt.bar(y_pos, results, align='center', alpha=0.5, width = 0.3, color = color)
    plt.xticks(y_pos, disk_names)
    plt.ylabel(y_label)
    plt.title(title)
    for i in range(len(results)):
        plt.text(x = y_pos[i] - 0.1 , y = results[i]+0.1, s = results[i], size = 8)
    plt.show()
```

```python
plot_bar(types, dollar_per_100MB, "Dollars per 100MB/s per Month", "Cost-efficiency for Different Kinds of Disks", 'g')
```
